## 真假流量所占的比例各有多少？

我们考虑一下僵尸用户的特性应该是什么?

最低级的僵尸用户, 可能没有用户名, 用的是基本的用户名

也可能是有一堆乱码构成

但是这样还不够实锤, 如果该用户注册后从来没有发过微博,考虑可能是假粉丝.

如果没有填写简介,转发之后完全没有人点赞,完全没有发过微博.

通过一番探索分析，我们把转发数据中转发者的关注或者粉丝数少于等于5、没有简介、转发之后被点赞数评论数再转发数都为0、微博会员等级为0级的数据，以及转发者的关注或者粉丝数大于等于5但昵称长“用户XXXXXXXX”这样的数据抽取了出来。

```python
data['user.screen_name']
```

```python
data_fake = data[((data['user.follow_count']<=5)|(data['user.followers_count']<=5))
     & (data['user.description']).isnull()
    &  (data['comments_count']==0)
    &  (data['attitudes_count']==0)
    &  (data['reposts_count']==0)
    &  (data['user.mbrank']==0)]
```

```python
# 随机抽样5个样本
data_fake.sample(5)
```

```python
data_fake.shape
```

```python
# 取出真粉的转发
data_true = data.drop(data_fake.index)
```

```python
data_true
```

```python
data_true.shape
```

```python
print('真粉丝转发数占总转发数的{}%'.format(np.round(data_true.shape[0]/data.shape[0]*100, 2)))
print('假粉丝转发数占总转发数的{}%'.format(np.round(data_fake.shape[0]/data.shape[0]*100, 2)))
```

```python
# 类的实例化,实例化一个柱状图的对象
bar = Bar(init_opts = opts.InitOpts(width = '550px',
    height = '450px'))

# 添加横轴和纵轴的数据
bar.add_xaxis(['总转发量', '假粉丝转发量', '真粉丝转发量'])
bar.add_yaxis("总数据102313条", [data.shape[0], data_fake.shape[0], data_true.shape[0]])
bar.set_global_opts(title_opts=opts.TitleOpts('蔡徐坤真假流量的转发量'))
bar.render_notebook()
```

天啦,假流量的比例居然这么高,102313条转发数据中，有95397条是由假粉丝转发的，占了总转发的93.24%，只有6916条是由真粉丝转发的，占6.76%.

这蔡徐坤水分也太大了吧
