## 手动实现KNN算法

1. 算距离,算出新的样本点距离之前的10个的距离
2. 确定一个k值, 从上面计算的距离中提取出最近的k个值

```python
new_data
```

```python
wine_data[['酒精浓度','颜色深度']]
```

```python
# 批量相减 (xa - xb)   (ya - yb)
wine_data[['酒精浓度','颜色深度']] - new_data
```

```python
# (xa - xb)**2   (ya - yb)**2
(wine_data[['酒精浓度','颜色深度']] - new_data) ** 2
```

```python
# 矢量化运算批量计算10个点的距离
distance = np.sqrt(np.sum((wine_data[['酒精浓度','颜色深度']] - new_data) ** 2, axis = 1))
distance
```

```python
# 确定一个k值 为 3
k = 3
```

```python
# 对原数据的索引值进行排序
distance.argsort()
```

```python
# 提取出前k个数据的索引值
distance.argsort()[:k]
```

```python
distance
```

```python
distance.argsort()[:k]
```

```python
distance[distance.argsort()[:k]]
```

```python
wine_data
```

```python
# k 个最近邻的分类值
k_vote = wine_data.loc[distance.argsort()[:k]]['品种']
k_vote
```

```python
k_vote.mode()[0]
```

```python
def KNN(wine_data ,new_data, k = 1 ):
    """KNN算法分类器:
    参数:wine_data : 训练数据集,输入数据
    new_data : 待分类数据
    k : 最近邻个数 
    返回值: 分类结果 0 或 1"""
    # 计算所有样本点距离
    # 矢量化运算批量计算10个点的距离
    distance = np.sqrt(np.sum((wine_data[['酒精浓度','颜色深度']] - new_data) ** 2, axis = 1))
    # 提取前k个最小值
    # k 个最近邻的分类值
    k_vote = wine_data.loc[distance.argsort()[:k]]['品种']
    # 提取出结果中的众数作为分类结果
    return k_vote.mode()[0]
```

## 绘制交叉验证学习曲线,找到最优的k值

```python
# 调用函数切割数据集
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3 , random_state = 100)
```

```python
# 建立空列表储存分数结果
scores = []
stds = []
# k 值 从1 到 30
for i in range(1,31):
    # 实例化模型
    knn = KNeighborsClassifier(n_neighbors= i)
    # 使用交叉验证
    cross_result = cross_val_score(knn, X_train , y_train , cv = 5)
    # 求平均值
    mean_score = cross_result.mean()
    # 求标准差
    std = cross_result.std()
    scores.append(mean_score)
    stds.append(std)
```

```python
# 找出最优的k值
best_k = np.argmax(scores) + 1
best_k
```
